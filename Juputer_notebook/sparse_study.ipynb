```python

#There is no need to call this function at all
#If the last two days downloaded are adject (it's actually a clever exercise to prove why this is true)


def generate_missing_google_trend_df_using_two_dfs_a_week_apart(left_df, right_df):
    #left_df, eg. a google trend score df containing dates from 10-15 -> 10-21 
    #right_df, eg. a google trend score df containing dates from 10-21 -> 10-27
    #and the goal is to generate every single missing df in between them

    #extract the keyword
    keyword = left_df.columns[1]

    #getting the date, and MOST important, finding the conversation ratio constant 
    overlap_time_df = pd.merge(left_df, right_df, how="inner", on="date")
    overlap_time = overlap_time_df["date"][0]
    std_conversion_ratio = np.mean(overlap_time_df[keyword+"_x"] / overlap_time_df[keyword+"_y"])
    print(std_conversion_ratio)



    outer_merge_df = pd.merge(left_df, right_df, how="outer", on="date")

    #this is every single day in short form in both left_df and right_df
    #it should look something like ['2021-10-15', '2021-10-16' ... '2021-10-27' ]
    the_13_days = outer_merge_df["date"].apply(lambda x : x.split(" ")[0]).unique()
    assert len(the_13_days) == 13, "the two dfs are not a week apart, violating function assumption, abort"

    #This is the df that's before the intersection of the two table
    #We want standardize the date's trend score into right_df measuring system
    before_df = outer_merge_df[outer_merge_df["date"] < overlap_time]
    std_trend_score_array = before_df[keyword + '_x']/std_conversion_ratio

    after_df = outer_merge_df[outer_merge_df["date"] >= overlap_time]
    og_trend_score_array = after_df[keyword + "_y"]
    
    entire_std_trend_score_array = pd.concat([std_trend_score_array, og_trend_score_array])

    #add to outer merge
    outer_merge_df["std_score"] = entire_std_trend_score_array


    #returns a dictonary of dates+dictionary pair
    #and we are generating 5 dfs, you can go count it
    result = {}
    for i in range(1,6):
        begin_date = the_13_days[i]
        df_name = the_13_days[i+6]
        end_date = the_13_days[i+7]
        df_i = outer_merge_df[(outer_merge_df["date"] >= begin_date) & (outer_merge_df["date"] <= end_date)][["date", "std_score"]]
        #clone is needed, since the above is only a view
        df_i = df_i.copy()
        df_i[keyword] = round((df_i["std_score"]/(max(df_i["std_score"])/100)), 0)
        df_i = df_i [["date", keyword]]
        result[df_name] = df_i

    return result

        







    
```
